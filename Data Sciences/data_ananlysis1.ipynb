- selecting rows and cols
  - rows (slicing)
     -'loc' - gets us the data using        named idexes
     -'iloc' - gets us the data using numeric indexes
  - cols (dictionary style)   
      - 'df[single] - requires name of column
      - 'df[multiple]- requires a list of column (multiple is list of named cols)

```python
# single row (record)

weather.iloc[0]
```

```python
weather.iloc[-1]
```

```python
weather.iloc[25:50]
```

```python
# records 1 - 15 with first 5 col

weather.iloc[1:16,:5]
```

```python
weather.iloc[1:16,[0,1,2,3,4]]
```

```python
# 15 records with last 3 cols

weather.iloc[:15,[-3,-2,-1]] 
```
